```
#Plotting of streamline of flow in circular cylinder

import numpy as np
import matplotlib.pyplot as plt
import sympy
from sympy.abc import x, y 

def cylinder_stream_function(U=10, R=1):
  r = sympy.sqrt(x**2 + y**2)
  theta = sympy.atan2(y, x)
  return U * (r - R**2 / r) * sympy.sin(theta)


def rotating_cylinder_stream_function(U=10, R=1,C=20):
  r = sympy.sqrt(x**2 + y**2)
  theta = sympy.atan2(y, x)
  return U * (r - R**2 / r) * sympy.sin(theta)+C*sympy.ln(r)/(2*np.pi*2*np.pi*R*R)

def uniform_flow_stream_function(U=10):
  r = sympy.sqrt(x**2 + y**2)
  theta = sympy.atan2(y, x)
  return U * r * sympy.sin(theta)

def velocity_field(psi):
  u = sympy.lambdify((x, y), psi.diff(y), 'numpy')
  v = sympy.lambdify((x, y), -psi.diff(x), 'numpy')
  return u, v

def plot_streamlines(ax, u, v, xlim=(-4, 4), ylim=(-4,4)):
    x0, x1 = xlim
    y0, y1 = ylim
    # create a grid of values
    Y, X =  np.ogrid[y0:y1:1000j, x0:x1:1000j]
    R = np.sqrt(X**2 + Y**2)

    u1=u(X,Y)
    v1=v(X,Y)

    u1[R<=1] = np.nan
    v1[R<=1] = np.nan
       


    return ax.streamplot(X, Y, u1, v1, color='cornflowerblue')


#plotting stream flow around a cylinder
psi = cylinder_stream_function()
u, v = velocity_field(psi)




fig, ax = plt.subplots(figsize=(5, 5))

plot_streamlines(ax, u, v,(-4,4),(-4,4))
plt.title('Streamline of flow in circular cylinder')
plt.show()
```
